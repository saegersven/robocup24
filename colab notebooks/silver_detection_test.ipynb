```python
from google.colab.patches import cv2_imshow
import numpy as np

NO_SILVER_PATH = '/content/no_silver'
SILVER_PATH = '/content/silver'

silver = []
no_silver = []

for filename in os.listdir(SILVER_PATH):
  img = cv2.imread(SILVER_PATH + '/' + filename)
  hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

  thresh = cv2.inRange(hsv, (0,0,0), (30,255,255))

  contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

  silver.append(cv2.countNonZero(thresh))

for filename in os.listdir(NO_SILVER_PATH):
  img = cv2.imread(NO_SILVER_PATH + '/' + filename)
  hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

  thresh = cv2.inRange(hsv, (0,0,0), (30,255,255))

  contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

  no_silver.append(cv2.countNonZero(thresh))

silver = np.array(silver)
no_silver = np.array(no_silver)

#print(f"{np.mean(silver)}  /  {np.min(silver)}  /  {np.max(silver)}")
#print(f"{np.mean(no_silver)}  /  {np.min(no_silver)}  /  {np.max(no_silver)}")

SILVER_TRES = 500
num_silver_detections = 0
num_no_silver_detections = 0

for i in silver:
  if i > SILVER_TRES:
    num_silver_detections = num_silver_detections + 1
print(num_silver_detections / len(silver), "(correct silver detections)")
for i in no_silver:
  if i < SILVER_TRES:
    num_no_silver_detections = num_no_silver_detections + 1
print(num_no_silver_detections / len(no_silver), "(correct no_silver detections)")
```
